```python
# Import the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a 104 X 104 matrix, A, and a 104 X 1 vector, x. Fill A and x with random numbers

A = np.random.rand(104, 104)
x = np.random.rand(104, 1)
```

```python
# Perform a matrix-vector multiplication

b = np.dot(A, x)
```

```python
# Create a scatter plot of the values in vector b

plt.scatter(range(len(b)), b, c='red')
plt.title("Scatter Plot of Vector b")
plt.xlabel("Index")
plt.ylabel("Value of b")
plt.grid(True)
plt.show()
```
